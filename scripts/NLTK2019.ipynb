```python
with open('/Users/13029/Desktop/output.csv', newline='', encoding='utf8') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
new_data = []

for i in range(len(data)):
    new_data.append(data[i][1].lower())
```

```python
cleaned_data = []

for i in range(len(new_data)):
    cleaned_data.append(new_data[i].translate(str.maketrans('', '', string.punctuation)))
```

```python
tokenized_words = []

for i in range(len(cleaned_data)):

    tokenized_words.append(word_tokenize(cleaned_data[i], "english"))
```

```python
for i in range(len(tokenized_words)):
    for word in tokenized_words[i]:
        if word in stopwords.words('english'):
            tokenized_words[i].remove(word)
```

```python
tok_final = tokenized_words
```

```python
for i in range(len(tok_final)):
    for word in tok_final[i]:
        word = WordNetLemmatizer().lemmatize(word)
   
```

```python
def sentiment_analyse(sentiment_text):
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    if score['neg'] > score['pos']:
        print("Negative Sentiment")
        return "Negative Sentiment"
    elif score['neg'] < score['pos']:
        print("Positive Sentiment")
        return "Positive Sentiment"
    else:
        print("Neutral Sentiment")
        return "Neutral Sentiment"
```

```python
post = []
sent = []
for i in range(len(cleaned_data)):
    post.append("post" + str(i + 1)) 
    sent.append(sentiment_analyse(cleaned_data[i]))
```

```python
df = pd.DataFrame(columns = ['Post #', 'Sentiment'])
```

```python
df['Post #'] = post
df['Sentiment'] = sent
```
